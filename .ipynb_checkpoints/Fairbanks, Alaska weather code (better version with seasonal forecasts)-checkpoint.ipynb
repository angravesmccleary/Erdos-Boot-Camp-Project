```python
## For data handling
from datetime import datetime
import pandas as pd
import numpy as np

## For plotting
import matplotlib.pyplot as plt
import seaborn as sns

## This sets the plot style
## to have a grid on a white background
sns.set_style("whitegrid") 
```

```python
#99709 is the ZIP code for Fairbanks, Alaska
alaska = pd.read_csv("99709.csv", parse_dates=['date_time'])


# This will tell us about the df
print("There are",len(alaska),"observations in df.")
print("The columns are",list(alaska.columns))

# Look at 5 randomly sampled rows
alaska.sample(5)
```

```python
alaska_train = alaska.loc[alaska.date_time < datetime(2018,1,1),].copy()
alaska_test = alaska.drop(alaska_train.index).copy()
```

```python
alaska_train.head(10)
```

```python
## first make a figure
## this makes a figure that is 8 units by 8 units
plt.figure(figsize = (8,8))

## plt.scatter plots RD on the x and W on the y
plt.scatter(alaska_train.tempC, alaska_train.totalSnow_cm, s=1)

## Always good practice to label well when
## presenting a figure to others
## place an xlabel
plt.xlabel("Temperature (Degrees C)", fontsize =16)

## place a ylabel
plt.ylabel("Total Snow (cm)", fontsize = 16)

## type this to show the plot
plt.show()
```

```python
## first make a figure
## this makes a figure that is 8 units by 8 units
plt.figure(figsize = (8,8))

## plt.scatter plots RD on the x and W on the y
plt.scatter(alaska_train.index.values, alaska_train.totalSnow_cm, s=1)

## Always good practice to label well when
## presenting a figure to others
## place an xlabel
plt.xlabel("Date", fontsize =16)

## place a ylabel
plt.ylabel("Total Snow (cm)", fontsize = 16)

## type this to show the plot
plt.show()
```

```python
from statsmodels.tsa.api import SimpleExpSmoothing
```

```python
ses = SimpleExpSmoothing(alaska_train.totalSnow_cm.values)
```

```python
fit = ses.fit(smoothing_level=0.5, optimized=False)
```

```python
# and the forecast on the test data
plt.figure(figsize=(16,12))

# plot the training data
plt.plot(alaska_train.index.values, alaska_train.totalSnow_cm,'b',
            label = "Training Data")

# plot the fit
plt.plot(alaska_train.index.values, fit.fittedvalues,'r-',
            label = "Fitted Values")

# plot the forecast
plt.plot(alaska_test.index.values, fit.forecast(len(alaska_test)),'r--',
            label = "Forecast")

plt.plot(alaska_test.index.values, alaska_test.totalSnow_cm,'b--',
            label = "Test Data")

plt.legend(fontsize=14)

plt.xlabel("Year", fontsize=16)
plt.ylabel("Production \n in millions of tonnes", fontsize=16)

plt.show()
```

```python

plt.figure(figsize=(16,12))

# plot the forecast
plt.plot(alaska_test.index.values, fit.forecast(len(alaska_test))-alaska_test.totalSnow_cm,'r--',
            label = "Forecast")
plt.legend(fontsize=14)

plt.plot(alaska_test.index.values, alaska_test.totalSnow_cm,'b--',
            label = "Test Data")
plt.xlabel("Year", fontsize=16)
plt.ylabel("Production \n in millions of tonnes", fontsize=16)

plt.show()
```

```python
#Now we add seasonality
from statsmodels.tsa.api import ExponentialSmoothing
#We remove the moon rise, moon set, sunrise, sunset
alaska2 = pd.read_csv("99709.csv", parse_dates=['date_time'])
alaska2.drop('moonrise', inplace=True, axis=1)
alaska2.drop('moonset', inplace=True, axis=1)
alaska2.drop('sunrise', inplace=True, axis=1)
alaska2.drop('sunset', inplace=True, axis=1)
alaska2.drop('date_time', inplace=True, axis=1)
alaska2_train = alaska2.loc[alaska.date_time < datetime(2018,1,1),].copy()
alaska2_test = alaska2.drop(alaska_train.index).copy()
```

```python
season_length = 181*8 #181 days between oct 1 and march 31, and 8 time steps a day
time_between_seasons = 365*8-season_length
add_mod = ExponentialSmoothing(alaska2_train.totalSnow_cm, 
                               seasonal_periods=time_between_seasons, 
                               seasonal='add')

# In the multiplicative model set seasonal to 'mul'
mul_mod = ExponentialSmoothing(alaska2_train.totalSnow_cm+1, #The +1 was added to ensure all values are strictly positive.
                               seasonal_periods=time_between_seasons, 
                               seasonal='mul')
```

```python
# The Additive Model
add_mod_fit = add_mod.fit()
mul_mod_fit = mul_mod.fit()
```

```python
fig,ax = plt.subplots(2,1,figsize=(12,16),sharex=True, sharey=True)

## The additive model
# training data
ax[0].plot(alaska2_train.index, alaska2_train.totalSnow_cm, 
           'b-', label="Training Data")

# fitted values
ax[0].plot(alaska2_train.index, add_mod_fit.fittedvalues, 
           'r-', label="Fitted Values")

# The test data
ax[0].plot(alaska2_test.index, alaska2_test.totalSnow_cm,
            'b--', label="Test Data")
ax[0].plot(alaska2_test.index, add_mod_fit.forecast(len(alaska2_test)),
           'r--', label="Forecast")

ax[0].legend(fontsize=14)

ax[0].set_title("Additive Model",fontsize=18)
ax[0].set_ylabel("Snowfall",fontsize=16)


#multipicative model
#Training data
ax[1].plot(alaska2_train.index, alaska2_train.totalSnow_cm+1, 
           'b-', label="Training Data")

# fitted values
ax[1].plot(alaska2_train.index, mul_mod_fit.fittedvalues, 
           'r-', label="Fitted Values")

# The test data
ax[1].plot(alaska2_test.index, alaska2_test.totalSnow_cm+1,
            'b--', label="Test Data")
ax[1].plot(alaska2_test.index, mul_mod_fit.forecast(len(alaska2_test)),
           'r--', label="Forecast")

ax[1].legend(fontsize=14)

ax[1].set_title("Multiplicative Model",fontsize=18)
ax[1].set_ylabel("Snowfall",fontsize=16)

plt.show()
```

```python
fig,ax = plt.subplots(1,1,figsize=(12,8),sharex=True, sharey=True)

## The additive model
# training data
ax.plot(alaska2_train.index, alaska2_train.totalSnow_cm, 
           'b-', label="Training Data")

# fitted values
ax.plot(alaska2_train.index, add_mod_fit.fittedvalues, 
           'r-', label="Fitted Values")

# The test data
ax.plot(alaska2_test.index, alaska2_test.totalSnow_cm,
            'b--', label="Test Data")
ax.plot(alaska2_test.index, add_mod_fit.forecast(len(alaska2_test)),
           'r--', label="Forecast")

ax.legend(fontsize=14)

ax.set_title("Additive Model",fontsize=18)
ax.set_ylabel("Snowfall",fontsize=16)

plt.show()
```

```python
fig,ax = plt.subplots(1,1,figsize=(12,8),sharex=True, sharey=True)

## The additive model
# training data
ax.plot(alaska2_train.index, alaska2_train.totalSnow_cm-add_mod_fit.fittedvalues, 
           'b-', label="Training Error")


# The test data
ax.plot(alaska2_test.index, alaska2_test.totalSnow_cm- add_mod_fit.forecast(len(alaska2_test)),
            'b--', label="Forecast Error")


ax.legend(fontsize=14)

ax.set_title("Additive Model",fontsize=18)
ax.set_ylabel("Snowfall",fontsize=16)

plt.show()
```
